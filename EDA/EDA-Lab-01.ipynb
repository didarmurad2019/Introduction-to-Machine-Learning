```python
# ===================Author Dr Didar Murad PhD (CS&E) =========================
# BS LEVEL PYTHON BASICS FOR MACHINE LEARNING
# Exploratory Data Analysis (EDA) is the process of understanding, cleaning, and summarizing data before applying machine learning models.
# EDA lab-01 (Zero level practice)
# ============================================
# Import pandas library and give it the short name 'pd'
import pandas as pd
# Import matplotlib pyplot module for plotting graphs
import matplotlib.pyplot as plt
```

```python
# Read the CSV file and store the data in a DataFrame called df
df = pd.read_csv("basic_eda_dataset.csv")
```

```python
print(df)
```

```python
# Print the first 5 rows of the dataset to quickly inspect the data
print(df.head())
```

```python
# Print the shape of the dataset (number of rows, number of columns)
print(df.shape)
```

```python
# Print information about the dataset such as data types and missing values
print(df.info())
```

```python
# Print statistical summary of numerical columns (mean, std, min, max)
print(df.describe())
```

```python
# Check and print the number of missing values in each column
print(df.isnull().sum())
```

```python
# Count and print the number of samples in each class of the target variable 'label'
print(df["label"].value_counts())
```

```python
# Draw a histogram to show the distribution of the 'salary' column
df["salary"].hist()
# Display the histogram plot
plt.show()
```

```python
# Create a scatter plot to show the relationship between age and salary
plt.scatter(df["age"], df["salary"])
plt.show()
```
